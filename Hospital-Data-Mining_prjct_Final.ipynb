# **Task 2 — Data Preprocessing**

## 1. Handle Missing Values



**Handling Missing Values:** We used SimpleImputer to fill missing features.

**Numeric columns:** Use median imputation (robust against outliers).

**Categorical columns:** Use most frequent imputation.

```python
from sklearn.impute import SimpleImputer

# Numeric
num_cols = df.select_dtypes(include="number").columns
num_imputer = SimpleImputer(strategy="median")
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Categorical
cat_cols = df.select_dtypes(include="object").columns
cat_imputer = SimpleImputer(strategy="most_frequent")
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

## 4. Detect & Treat Outliers

Use IQR method or z-score for numeric features.

```python
num_cols = df.select_dtypes(include="number").columns

def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    return outliers

# Count outliers per column
outlier_counts = {}
for col in num_cols:
    outlier_counts[col] = detect_outliers_iqr(df, col).shape[0]

# Convert to DataFrame for easy viewing
outlier_summary = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Number of Outliers'])
print(outlier_summary)
```

# 1. Elbow Method

The Elbow Method was used to determine the optimal value of **K** by measuring the clustering cost (inertia) for different numbers of clusters. As the number of clusters increases, the cost decreases. However, after a certain point, the improvement becomes smaller. This point forms an “elbow” shape in the graph and indicates the most appropriate number of clusters.

**Why we used it:**
- To avoid choosing too few or too many clusters.
- To find a balance between model simplicity and clustering quality.
- To identify the point where adding more clusters provides minimal improvement

```python
costs = []

for k in range(2, 8):
    model = KMedoids(n_clusters=k, method='alternate',random_state=42)
    model.fit(sample_df)
    costs.append(model.inertia_)

plt.plot(range(2, 8), costs, marker='o' , color="#9777C2")
plt.title("Elbow Method (K-Medoids)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Cost (Inertia)")
plt.show()
```

# 2. Silhouette Score

The Silhouette Score was used to evaluate how well each data point fits within its assigned cluster compared to other clusters. A higher score indicates that clusters are compact and clearly separated.

**Why we used it:**
- To measure the quality of clustering results.
- To ensure that clusters are distinct and internally consistent.
- To validate the value of **K** selected from the Elbow Method.

```python

scores = []

for k in range(2, 8):
    model = KMedoids(n_clusters=k, method='alternate', random_state=42)
    labels = model.fit_predict(sample_df)

    score = silhouette_score(sample_df, labels)
    scores.append(score)

    print(f"K = {k}, Silhouette Score = {score:.4f}")

plt.plot(range(2, 8), scores, marker='o' , color ="#9777C2")
plt.title("Silhouette Score (K-Medoids)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Score")
plt.show()
```

### Step 3 — **Train the Final Model**

Based on the Elbow and Silhouette results, we select the optimal k and train the final K-Medoids model.
Each patient is assigned a cluster label, which we add as a new column in the dataset.

### Choosing the Optimal K

After removing non-clinical features (blood type, gender, insurance provider),
the Elbow Method suggested k=3, where the inertia curve begins to flatten.
The Silhouette Score was highest at k=2 (0.56), while k=3 achieved a score of 0.46.

Since both methods disagreed, we selected k=3 based on the Elbow Method,
as the Silhouette Score difference is marginal and k=3 provides a richer
clinical interpretation (mild / moderate / severe cases).

```python
best_k = 3

model = KMedoids(n_clusters=best_k, method='alternate', random_state=42)
model.fit(sample_df)

sample_df = sample_df.copy()
sample_df['Cluster'] = model.labels_
```

# Task 4 — Hierarchical Clustering

Hierarchical Clustering is an unsupervised algorithm that builds a tree-like
structure (dendrogram) to group similar data points together.
Unlike K-Medoids, it does not require specifying k in advance —
the dendrogram helps us visually decide the optimal number of clusters.

We apply three linkage methods:
- **Ward:** Minimizes the total variance within clusters
- **Complete:** Uses the maximum distance between points in two clusters
- **Average:** Uses the mean distance between all point pairs across clusters

We then use the Elbow Method and Silhouette Score to confirm the optimal k,
and interpret each cluster in the healthcare domain context.

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
```

### Step 1 — Dendrogram

We plot the dendrogram for each linkage method to visually inspect
how the data groups together and get an initial idea of the optimal k.
We use a sample of 4000 points for visualization clarity.

```python
# use small sample for dendrogram visualization only
dendro_sample = sample_df.drop(columns=['Cluster']).sample(n=4000, random_state=42)

linkage_methods = ['ward', 'complete', 'average']
colors = ['#9B59B6', '#5DADE2', '#48C9B0']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, method, color in zip(axes, linkage_methods, colors):
    Z = linkage(dendro_sample, method=method)
    dendrogram(Z, ax=ax, truncate_mode='lastp', p=20,
               color_threshold=0, above_threshold_color=color)
    ax.set_title(f'Dendrogram — {method.capitalize()} Linkage')
    ax.set_xlabel('Data Points')
    ax.set_ylabel('Distance')

plt.tight_layout()
plt.show()
```

### Step 2 — Elbow Method
We run Agglomerative Clustering for k values from 2 to 7
across all three linkage methods and record the inertia each time.
The point where the curve begins to flatten indicates the optimal k

```python
# drop both cluster columns if they exist
cols_to_drop = [c for c in ['Cluster', 'HC_Cluster'] if c in sample_df.columns]
X = sample_df.drop(columns=cols_to_drop).values

linkage_methods = ['ward', 'complete', 'average']
colors = ['#9B59B6', '#5DADE2', '#48C9B0']

inertia_dict = {method: [] for method in linkage_methods}
k_range = range(2, 7)

for method in linkage_methods:
    for k in k_range:
        model = AgglomerativeClustering(n_clusters=k, linkage=method)
        labels = model.fit_predict(X)
        inertia = sum(
            np.sum((X[labels == i] - X[labels == i].mean(axis=0)) ** 2)
            for i in range(k)
        )
        inertia_dict[method].append(inertia)

plt.figure(figsize=(8, 4))
for method, color in zip(linkage_methods, colors):
    plt.plot(k_range, inertia_dict[method], marker='o', label=method.capitalize(), color=color)
plt.title('Elbow Method — Hierarchical Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Choosing the Optimal K

The Elbow Method showed no sharp elbow across all linkage methods,
reflecting the uniform feature distribution in this dataset.
However, Complete linkage showed the clearest inflection point at k=3,
where the rate of inertia decrease slows noticeably compared to Ward and Average.
We therefore selected k=3 as the optimal number of clusters.

### Step 3 — Silhouette Score

We calculate the Silhouette Score for each linkage method across k values
to confirm which k produces the most well-separated clusters.

```python
sil_dict = {method: [] for method in linkage_methods}

for method in linkage_methods:
    for k in k_range:
        model = AgglomerativeClustering(n_clusters=k, linkage=method)
        labels = model.fit_predict(X)
        sil_dict[method].append(silhouette_score(X, labels))
        print(f"{method.capitalize()} | K = {k}, Silhouette Score = {sil_dict[method][-1]:.4f}")

plt.figure(figsize=(8, 4))
for method, color in zip(linkage_methods, colors):
    plt.plot(k_range, sil_dict[method], marker='o', label=method.capitalize(), color=color)
plt.title('Silhouette Score — Hierarchical Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Why Single Linkage Was Excluded

Single linkage was initially tested but excluded from the final analysis.
It suffered from the **chaining effect** — where most data points were
assigned to one large cluster while the remaining clusters contained
very few points. This resulted in a flat, non-informative inertia curve
and a negative Silhouette Score, making it unsuitable for this dataset.

Ward and Complete linkage produced more balanced and meaningful clusters,
and were therefore selected for the final analysis.

### Step 4 — Train the Final Model

Based on the Elbow and Silhouette results, we select the best linkage method
and optimal k to train the final Hierarchical Clustering model.

### Choosing the Optimal K and Linkage Method

The Elbow Method showed no sharp elbow across all linkage methods,
reflecting the uniform feature distribution in this dataset.
However, the rate of inertia decrease slows most noticeably at k=3,
suggesting it as the optimal number of clusters.

The Silhouette Score was highest at k=2 for all methods.
However, since the difference between k=2 and k=3 is marginal,
and k=3 provides a clinically richer interpretation (mild / moderate / severe),
we selected k=3 as the optimal number of clusters.

At k=3, Complete linkage achieved the highest Silhouette Score (0.4489)
compared to Ward (0.4286) and Average (0.4023), making it the most
internally consistent clustering. Complete linkage was therefore selected
as the final method.

```python
best_k = 3
best_linkage = 'complete'

hc_model = AgglomerativeClustering(n_clusters=best_k, linkage=best_linkage)
sample_df['HC_Cluster'] = hc_model.fit_predict(X)

hc_labels = hc_model.labels_
hc_centroids = np.array([
    X[hc_labels == k].mean(axis=0)
    for k in range(best_k)
])
hc_model.cluster_centers_ = hc_centroids
print("HC centroids saved:", hc_centroids.shape)

print(f"Cluster distribution:")
print(sample_df['HC_Cluster'].value_counts().sort_index())
```

# *Fuzzy Rules Logic: Retrospective Burden Assessment*
Since the Severity is derived from the actual Length of Stay and the Billing Amount is already finalized, these rules serve as a retrospective audit to determine how much financial relief a patient deserved based on the "Resource Consumption" during their visit:

# Low Resource Consumption (Mild Severity):

If the patient had a short stay (Mild) and a Low/Medium bill, they are considered to have had a "Routine Visit" that doesn't justify financial aid (Rules 1 & 2).

If the stay was short but the bill was unexpectedly High, they receive Partial support (Rule 3).

# Average Resource Consumption (Moderate Severity):

For patients with average stays, support only kicks in if the bill exceeds the Low threshold (Rules 5 & 6). This balances the aid between clinical duration and actual cost.

# High Resource Consumption (Severe Severity):

Patients with the longest stays (Severe) are prioritized for support because long hospitalization significantly impacts their socio-economic status.

Even if the bill was Low, they get Partial aid for the "Time Burden" (Rule 7). If the bill was Medium or High, they qualify for Full support (Rules 8 & 9).

# **Task 6 — Genetic Algorithm Optimization**

**Purpose & Motivation**

In Tasks 3 and 4, we applied K-Medoids and Hierarchical Clustering using all available features. While both converged on k=3 clusters differentiated by length of stay, it is unknown whether all features contribute equally — or whether a smaller subset could produce equally good clustering with less noise.
A Genetic Algorithm is well-suited here because:

The feature selection search space is combinatorial (2ⁿ subsets for n features)
Exhaustive search is infeasible
GA intelligently evolves candidate solutions toward higher fitness

**Goal**: Find the optimal feature subset that maximizes the Silhouette Score of K-Medoids (k=3), and compare to the baseline (all features).

**GA Component Definitions**

*   Chromosome: Binary vector of length = number of features. One candidate feature subset.

*   Gene: A single bit `1` = feature included , `0` = excluded
*   Population: A set of chromosomes evaluated each generation.
*   Fitness Function: Silhouette Score of K-Medoids (k=3) on the selected features.
*   Selection: Tournament — two random chromosomes compete, fitter one is selected as parent.
*   Crossover: Single-point — a random cut splits two parents; halves are swapped to form two children.
*   Mutation: Bit-flip — each gene flips with a small probability to maintain diversity.
*   Children: Offspring produced by crossover + mutation; they replace the old population.

**Step 1 — Reconstruct Feature Matrix**

```python
import warnings, random
import numpy as np
warnings.filterwarnings('ignore')

# Remove cluster label columns added by Tasks 3 & 4
cols_to_drop = [c for c in ['Cluster' , 'HC_Cluster'] if c in sample_df.columns]
X_all = sample_df.drop(columns=cols_to_drop).values
feature_names = sample_df.drop(columns=cols_to_drop).columns.tolist()
n_features = len(feature_names)


# Baseline score — reuse the silhouette score already computed in Task 3
baseline_score = silhouette_score(X_all, model.labels_)

#use a fixed subsample for fitness evaluation
GA_SAMPLE_SIZE = 3000
rng = np.random.default_rng(42)
ga_idx = rng.choice(len(X_all), size=GA_SAMPLE_SIZE, replace=False)
X_ga = X_all[ga_idx]

print(f"Feature Space: {n_features} features")
print(f"GA Sample Size: {GA_SAMPLE_SIZE} / {len(X_all)} patients")
print(f"Features: {feature_names}")
```

**Step 2 — Fitness Function**

The fitness function takes a binary chromosome and returns the Silhouette Score of K-Medoids using only the features where the gene = 1.

**Why Silhouette Score?**

* Measures both cluster compactness and separation simultaneously
* Requires no ground truth labels — ideal for unsupervised optimization
* Ranges from -1 (bad) to +1 (perfect), making it interpretable

A penalty of `0.0` is returned when fewer than 2 features are selected, since clustering is undefined on a single feature.

```python
def fitness(chromosome, X=X_ga, k=best_k):
    selected = [i for i, gene in enumerate(chromosome) if gene == 1]
    if len(selected) < 2:
      return 0.0

    X_selected = X[:, selected]
    model = KMedoids(n_clusters=k, method='alternate', random_state=42, max_iter=100)
    labels = model.fit_predict(X_selected)

    if len(set(labels)) < 2:
        return 0.0
    return silhouette_score(X_selected, labels)
```

**Step 3 — GA Operators**

* Initialization: Each chromosome is randomly generated — every gene is independently set to 0 or 1 with equal probability.
* Tournament Selection: Two chromosomes are randomly picked; the fitter one becomes a parent. Repeated to select both parents per crossover.
* Single-Point Crossover: A random cut point is chosen. Child 1 = Parent 1's genes before cut + Parent 2's genes after. Child 2 is the reverse.
* Bit-Flip Mutation: Each gene has a small probability (`mutation_rate`) of flipping, preventing premature convergence.

```python
def initialize_population(pop_size, n_features):
    return [[random.randint(0,1) for _ in range (n_features)] for _ in range(pop_size)]

def select_parent(population, fitness_scores):
    a, b = random.sample(range(len(population)), 2)
    winner = a if fitness_scores[a] >= fitness_scores[b] else b
    return population[winner][:]

def crossover(parent1, parent2):
    point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

def mutate(chromosome, mutation_rate):
    for i in range(len(chromosome)):
        if random.random() < mutation_rate:
            chromosome[i] = 1 - chromosome[i]
    return chromosome
```

**Step 4 — GA Main Loop**

Each generation:

1. Evaluate fitness for every chromosome
2. Elitism — carry the best chromosome unchanged into the next generation
3. Build new population via tournament selection → crossover → mutation
4. Track best fitness across all generations

**Hyperparameter choices:**

* `pop_size = 30` : enough diversity without being slow
* `n_generations = 20` : sufficient for convergence on this feature space
* `crossover_rate = 0.8` : high rate encourages exploration
* `mutation_rate = 0.05` : low enough to not destroy good solutions

```python
pop_size = 20
n_generations = 15
crossover_rate = 0.8
mutation_rate = 0.05

np.random.seed(42)

population = initialize_population(pop_size, n_features)
best_fitness = -1
best_chromosome = None
generation_bests = []
generation_avgs = []

#Evaluting initial population
print("Evaluating initial population...")
initial_scores = [fitness(chrom) for chrom in population]
initial_best = max(initial_scores)
print(f"Initial population — Best Silhouette Score : {initial_best:.4f}")
print(f"Initial population — Avg  Silhouette Score : {np.mean(initial_scores):.4f}")


#Evolve loop
print("\n Running GA")
for gen in range(n_generations):

    fitness_score = [fitness(chrom) for chrom in population]

    gen_best_score = max(fitness_score)
    gen_best_chrom = population[fitness_score.index(gen_best_score)]
    generation_bests.append(gen_best_score)
    generation_avgs.append(np.mean(fitness_score))

    if gen_best_score > best_fitness:
        best_fitness = gen_best_score
        best_chromosome = gen_best_chrom[:]

    print(f"  Generation {gen+1:02d} | Best: {gen_best_score:.4f} | Avg: {np.mean(fitness_score):.4f}")


    new_population = [best_chromosome[:]]


    while len(new_population) < pop_size:
        parent1 = select_parent(population, fitness_score)
        parent2 = select_parent(population, fitness_score)

        if random.random() < crossover_rate:
            child1, child2 = crossover(parent1, parent2)
        else:
            child1, child2 = parent1[:], parent2[:]

        child1 = mutate(child1, mutation_rate)
        child2 = mutate(child2, mutation_rate)

        new_population.append(child1)
        if len(new_population) < pop_size:
            new_population.append(child2)

    population = new_population

print(f"\nGA complete. Best Silhouette Score found: {best_fitness:.4f}")
# Validate best chromosome on full data
selected_idx = [i for i, gen in enumerate(best_chromosome) if gen == 1]
final_score = silhouette_score(X_all[:, selected_idx],
                               KMedoids(n_clusters=best_k, method='alternate', random_state=42)
                              .fit_predict(X_all[:, selected_idx]))

print(f"Validated on full 31k rows    : {final_score:.4f}")
```

# **Task 7-System Implementation**

 Step 1 — Data Preparation (prepare_patient_data)
Converts a raw patient dict into a model-ready feature matrix. It calculates length of stay from admission/discharge dates, one-hot encodes categoricals, aligns columns to the training schema, and optionally scales numerics.

 Step 2 — Patient Classification
Runs two clustering models (K-Medoids and Hierarchical) on the prepared data. Cluster numbers (0/1/2) map to Mild / Moderate / Severe severity buckets.

 Step 3 — Financial Support via Fuzzy Logic
Feeds the severity score and billing amount into a fuzzy system that outputs a support percentage, then buckets it into Minimal / Partial / Full support.

 Step 4 — Genetic Algorithm Insights
Reads pre-computed GA results to identify which features matter most, then looks up that patient's values for those features.

  Step 5 — Composite Risk Score
Combines severity, billing, and length-of-stay into a 0–100 risk score using equal thirds (33.33 points each).

 Step 6 — Final Report
Assembles everything into a structured dict and prints a formatted summary.

```python
def prepare_patient_data(record: dict, preprocessors: dict):
    """Clean the raw record and get it ready for the models."""

    patient_df = pd.DataFrame([record]) if isinstance(record, dict) else record.copy()

    # Clean up column names
    patient_df.columns = (
        patient_df.columns.str.strip().str.lower().str.replace(" ", "_")
    )

    # Calculate stay from dates if we have them, otherwise fall back
    if "date_of_admission" in patient_df.columns and "discharge_date" in patient_df.columns:
        patient_df["date_of_admission"] = pd.to_datetime(patient_df["date_of_admission"])
        patient_df["discharge_date"]    = pd.to_datetime(patient_df["discharge_date"])
        computed_stay = int(
            (patient_df["discharge_date"] - patient_df["date_of_admission"]).dt.days.iloc[0]
        )
        patient_df["length_of_stay"] = computed_stay
    elif "length_of_stay" in patient_df.columns:
        computed_stay = int(patient_df["length_of_stay"].iloc[0])
    else:
        computed_stay = 14
        patient_df["length_of_stay"] = computed_stay

    # Drop columns the models don't need
    remove_cols = ["name", "doctor", "hospital", "room_number",
                   "date_of_admission", "discharge_date"]
    patient_df = patient_df.drop(columns=[c for c in remove_cols if c in patient_df.columns])

    # Fill nulls
    num_cols_hint = preprocessors.get("num_cols", [])
    for col in patient_df.columns:
        if patient_df[col].isnull().any():
            patient_df[col] = patient_df[col].fillna(0 if col in num_cols_hint else "Unknown")

    # One-hot encode all text columns
    for col in patient_df.select_dtypes(include=["object"]).columns:
        dummies = pd.get_dummies(patient_df[col], prefix=col)
        patient_df = pd.concat([patient_df.drop(columns=[col]), dummies], axis=1)

    # Scale — add dummy room_number=0 if scaler expects it, then drop after
    if preprocessors.get("scaler") is not None:
        cols_to_scale = list(preprocessors["scaler"].feature_names_in_)
        for col in cols_to_scale:
            if col not in patient_df.columns:
                patient_df[col] = 0
        patient_df[cols_to_scale] = preprocessors["scaler"].transform(patient_df[cols_to_scale])

    # Align to training schema
    expected_cols = preprocessors.get("feature_names", patient_df.columns.tolist())
    for col in expected_cols:
        if col not in patient_df.columns:
            patient_df[col] = 0
    patient_df = patient_df[expected_cols]

    return patient_df, computed_stay


def extract_billing_amount(record: dict) -> float:
    """Pull billing amount from the record. Warn if it's missing."""
    raw = record.get("billing_amount")

    if raw is None:
        print("Warning: billing_amount is missing. Financial assessment will be unreliable.")
        return 0.0

    if isinstance(raw, str):
        return float(raw.replace("$", "").replace(",", ""))

    return float(raw)


def hierarchical_predict(hierarchical_model, patient_features: np.ndarray) -> int:
    """
    Find the nearest cluster centroid for this patient.
    Re-fitting on one row is meaningless, so we do a distance lookup instead.
    """
    if not hasattr(hierarchical_model, "cluster_centers_"):
        raise AttributeError(
            "hierarchical_model needs a cluster_centers_ attribute. "
            "Save the centroids when you train the model."
        )

    distances = np.sum((hierarchical_model.cluster_centers_ - patient_features) ** 2, axis=1)
    return int(np.argmin(distances))

KMEDOIDS_WEIGHT       = 0.5
HIERARCHICAL_WEIGHT   = 0.5
MAX_BILLING_THRESHOLD = 55000

def healthcare_analytics_system(
    kmedoids_model,
    hierarchical_model,
    fuzzy_system,
    genetic_results: dict,
    preprocessing_pipeline: dict,
    patient_record: dict,
) -> dict:
    """
    Run a full clinical and financial analysis on one patient record.
    Returns a dict with all findings and recommendations.
    """

    print("  HEALTHCARE ANALYTICS SYSTEM — PATIENT ANALYSIS REPORT")

    cluster_meanings = {
        0: {
            "severity": "Mild",
            "description": "Short hospital stay, routine recovery",
            "risk": "Low",
            "action": "Standard care plan",
        },
        1: {
            "severity": "Moderate",
            "description": "Average hospital stay, needs monitoring",
            "risk": "Medium",
            "action": "Enhanced care and regular checkups",
        },
        2: {
            "severity": "Severe",
            "description": "Long hospital stay, serious condition",
            "risk": "High",
            "action": "Intensive care and close monitoring",
        },
    }

    # Step 1 — Prepare the data
    print("\n[Step 1] Processing Patient Data")

    # computed_stay comes back here so we never read it from the raw record again
    prepared_patient, computed_stay = prepare_patient_data(
        patient_record, preprocessing_pipeline
    )
    print(f"  Data ready  |  Length of stay: {computed_stay} day(s)")

    # Step 2 — Classify the patient using both models
    print("\n[Step 2] Patient Classification")

    kmedoids_cluster = int(kmedoids_model.predict(prepared_patient)[0])

    # Nearest-centroid lookup instead of re-fitting on a single row
    hierarchical_cluster = hierarchical_predict(hierarchical_model, prepared_patient.values)

    # Combine both results, weighted by how reliable each method is
    raw_ensemble = KMEDOIDS_WEIGHT * kmedoids_cluster + HIERARCHICAL_WEIGHT * hierarchical_cluster
    ensemble_cluster = int(round(raw_ensemble))

    patient_info = cluster_meanings.get(
        ensemble_cluster,
        {
            "severity": "Unknown",
            "description": "Cannot classify",
            "risk": "Unknown",
            "action": "Need further assessment",
        },
    )

    print(f"  K-Medoids cluster    : {kmedoids_cluster}")
    print(f"  Hierarchical cluster : {hierarchical_cluster}")
    print(f"  Ensemble result      : {ensemble_cluster}")
    print(f"  Severity             : {patient_info['severity']}")
    print(f"  Condition            : {patient_info['description']}")
    print(f"  Risk level           : {patient_info['risk']}")
    print(f"  Suggested action     : {patient_info['action']}")

    severity_score = ensemble_cluster

    # Step 3 — Work out financial support
    print("\n[Step 3] Financial Support Calculation")

    billing_amount = extract_billing_amount(patient_record)
    support_percent = fuzzy_system.calculate_support(severity_score, billing_amount)

    if support_percent < 20:
        support_level = "Minimal Support"
        support_advice = "Patient can likely manage costs independently"
    elif support_percent < 60:
        support_level = "Partial Support"
        support_advice = "Recommend partial financial assistance"
    else:
        support_level = "Full Support"
        support_advice = "Maximum financial aid recommended"

    print(f"  Total bill   : ${billing_amount:,.2f}")
    print(f"  Support      : {support_percent:.1f}%")
    print(f"  Category     : {support_level}")
    print(f"  Advice       : {support_advice}")

    # Step 4 — Show what the genetic algorithm found
    print("\n[Step 4] Feature Optimisation Results")

    best_features = genetic_results.get("optimal_features", [])
    quality_boost = genetic_results.get("improvement", 0)

    print(f"  Optimal features : {len(best_features)}")
    print(f"  Quality boost    : {quality_boost:.1%}")

    if best_features:
        print("\n  Top clinical factors:")
        for idx, feature in enumerate(best_features[:5]):
            value = patient_record.get(feature, "Not available")
            print(f"    {idx + 1}. {feature}: {value}")

    # Step 5 — Build the composite risk score
    print("\n[Step 5] Overall Risk Assessment")

    # Three equal components, each worth up to 33.33 points
    severity_risk = severity_score * 33.33
    billing_risk = min(billing_amount / MAX_BILLING_THRESHOLD, 1.0) * 33.33

    # Use computed_stay here, not the raw record value
    if computed_stay > 30:
        stay_risk = 33.33
    elif computed_stay > 14:
        stay_risk = 16.67
    else:
        stay_risk = 0.0

    risk_total = severity_risk + billing_risk + stay_risk

    if risk_total < 33.33:
        risk_group = "Low Risk"
        final_plan = "Standard discharge, routine follow-up scheduled"
    elif risk_total < 66.67:
        risk_group = "Moderate Risk"
        final_plan = "Close monitoring needed, consider social work consultation"
    else:
        risk_group = "High Risk"
        final_plan = "Immediate care coordination, financial counselling required"

    print(f"  Severity component : {severity_risk:.1f} / 33.33")
    print(f"  Billing component  : {billing_risk:.1f} / 33.33")
    print(f"  Stay component     : {stay_risk:.1f} / 33.33")
    print(f"  Total risk score   : {risk_total:.1f} / 100")
    print(f"  Risk category      : {risk_group}")
    print(f"  Plan               : {final_plan}")

    # Step 6 — Package everything up
    print("  FINAL PATIENT ANALYSIS REPORT")

    final_results = {
        "patient_name": patient_record.get("name", patient_record.get("patient_id", "Unknown")),
        "age": patient_record.get("age", "Unknown"),
        "gender": patient_record.get("gender", "Unknown"),
        "medical_condition": patient_record.get("medical_condition", "Unknown"),
        "kmedoids_cluster": kmedoids_cluster,
        "hierarchical_cluster": hierarchical_cluster,
        "ensemble_cluster": ensemble_cluster,
        "severity": patient_info["severity"],
        "risk_level": patient_info["risk"],
        "billing_amount": billing_amount,
        "financial_support_percent": round(support_percent, 1),
        "support_category": support_level,
        "computed_stay_days": computed_stay,
        "risk_score": round(risk_total, 1),
        "risk_category": risk_group,
        "recommended_action": final_plan,
        "clinical_plan": patient_info["action"],
        "financial_advice": support_advice,
        "ga_improvement_pct": round(quality_boost * 100, 1),
        "features_used": len(best_features),
    }

    print(f"""
PATIENT SUMMARY
---------------
Name             : {final_results['patient_name']}
Age              : {final_results['age']}   |   Gender: {final_results['gender']}
Medical Condition: {final_results['medical_condition']}

CLINICAL ASSESSMENT
-------------------
K-Medoids Cluster    : {final_results['kmedoids_cluster']}
Hierarchical Cluster : {final_results['hierarchical_cluster']}
Ensemble Cluster     : {final_results['ensemble_cluster']}
Severity             : {final_results['severity']}
Clinical Risk        : {final_results['risk_level']}
Care Plan            : {final_results['clinical_plan']}

FINANCIAL ASSESSMENT
--------------------
Total Bill       : ${final_results['billing_amount']:,.2f}
Support Level    : {final_results['financial_support_percent']}%
Support Type     : {final_results['support_category']}
Advice           : {final_results['financial_advice']}

RISK EVALUATION
---------------
Length of Stay   : {final_results['computed_stay_days']} day(s)
Risk Score       : {final_results['risk_score']} / 100
Risk Category    : {final_results['risk_category']}
Recommended Plan : {final_results['recommended_action']}

SYSTEM PERFORMANCE
------------------
GA Optimisation Gain : {final_results['ga_improvement_pct']}%
Optimal Features     : {final_results['features_used']}
""")

    print("Analysis complete.")

    return final_results


def demonstrate_system():
    """Quick demo using mock models and a sample patient."""

    sample_patient = {
        "name": "Sarah Johnson",
        "age": 72,
        "gender": "Female",
        "medical_condition": "Diabetes Type 2",
        "date_of_admission": "2024-01-10",
        "discharge_date": "2024-02-05",
        "billing_amount": 38_500,
        "admission_type": "Emergency",
        "medication": "Metformin",
        "test_results": "Borderline",
    }

    class MockKMedoids:
        def predict(self, X):
            return [2]

    class MockHierarchical:
        # Centroids stored at training time so inference works properly
        cluster_centers_ = np.array([
            [0.1, 0.1, 0.1],
            [0.5, 0.5, 0.5],
            [0.9, 0.9, 0.9],
        ])

    class MockFuzzy:
        def calculate_support(self, severity, billing):
            if severity == 2 and billing > 30_000:
                return 78.5
            return 35.0

    mock_genetics = {
        "optimal_features": ["age", "billing_amount", "length_of_stay", "medical_condition"],
        "improvement": 0.18,
    }

    mock_preprocessing = {
        "scaler": None,
        "feature_names": ["age", "billing_amount", "length_of_stay"],
        "num_cols": ["age", "billing_amount", "length_of_stay"],
    }

    results = healthcare_analytics_system(
        kmedoids_model=MockKMedoids(),
        hierarchical_model=MockHierarchical(),
        fuzzy_system=MockFuzzy(),
        genetic_results=mock_genetics,
        preprocessing_pipeline=mock_preprocessing,
        patient_record=sample_patient,
    )

    return results


if __name__ == "__main__":
    print("Running demo\n")
    demo_results = demonstrate_system()
```

# Conclusion

1. **EXPLORATORY DATA ANALYSIS**
   The healthcare dataset (55,500 patients, 15 features) revealed a
   uniformly distributed population across medical conditions, admission
   types, and insurance providers. Age spans the full adult range with no
   dominant group. Billing amounts are approximately normally distributed
   with no extreme skew. The most clinically significant engineered feature
   was Length of Stay, derived from admission and discharge dates, which
   later proved to be the primary differentiator in clustering.

2. **PREPROCESSING**
   No missing values existed in the raw data. Negative billing values were
   clipped. Duplicate rows were removed. High-cardinality identifiers
   (name, doctor, hospital) were excluded. Seven categorical columns were
   one-hot encoded and all numeric features were standardised with
   StandardScaler to ensure distance-based algorithms were not biased by
   feature scale.

3. **K-MEDOIDS CLUSTERING**
   K=3 was selected via the Elbow Method (clearest inertia inflection) and
   supported by the Silhouette Score (0.46 at k=3 vs 0.56 at k=2 — a
   marginal difference that does not justify losing clinical granularity).
   The three clusters map cleanly onto patient severity:
     • Cluster 1 (Mild)     — avg. stay ~6 days, routine recovery
     • Cluster 0 (Moderate) — avg. stay ~16 days, standard monitoring
     • Cluster 2 (Severe)   — avg. stay ~26 days, extended critical care
   All other features (age, billing, condition type) were uniformly
   distributed across clusters, confirming length of stay as the sole
   differentiating signal — a clinically meaningful finding.

4. **HIERARCHICAL CLUSTERING**
   Ward, Complete, and Average linkage were all tested. Complete linkage
   with k=3 achieved the highest Silhouette Score (0.449) and was selected
   as the final method. The resulting clusters matched the K-Medoids groups
   almost exactly (mild/moderate/severe by stay duration), validating that
   the patient severity pattern is robust across algorithms and is not an
   artifact of any single method. Single linkage was excluded due to the
   chaining effect producing one dominant cluster.

5. **FUZZY LOGIC SYSTEM**
   A two-input fuzzy inference system was designed using cluster severity
   and billing amount to compute a financial support percentage. Nine
   IF-THEN rules grounded in healthcare resource logic were defined,
   covering all combinations of low/medium/high billing and mild/moderate/
   severe clinical severity. Centroid defuzzification produced a continuous
   support score (0–100%). Validation on five real patients confirmed
   correct behaviour: severe patients with high bills received ~88–90%
   support, while moderate patients with low bills received ~11–13%.
   Business value: this system automates fair, explainable financial aid
   allocation — replacing binary yes/no decisions with proportional,
   auditable recommendations.

6. **GENETIC ALGORITHM**
   A GA was applied to feature subset selection, optimising the Silhouette
   Score of K-Medoids (k=3) across 20 generations with a population of 30.
   The GA identified a reduced feature set that improved or matched the
   baseline clustering quality while using fewer features, reducing model
   complexity. This demonstrates that many one-hot encoded columns are
   redundant for clustering purposes, and that length-of-stay-related
   features carry the majority of the discriminative signal.

7. **INTEGRATED SYSTEM**
   The healthcare_analytics_system() pipeline accepts a raw patient record
   and produces a full report covering clinical classification (via
   K-Medoids), financial support recommendation (via Fuzzy Logic), and
   risk scoring. The GA-identified optimal feature set informs which patient
   attributes matter most. Together, the system enables:
     • Resource allocation based on predicted patient severity
     • Automated, explainable financial aid decisions
     • Early identification of high-risk patients for intervention

**OVERALL RESEARCH VALUE**
   This analysis shows that even in a dataset with uniform feature
   distributions, meaningful patient stratification is achievable through
   length of stay alone. Hospitals can use these findings to design
   tiered care pathways, target financial counselling to severe cases, and
   reduce administrative overhead through automated support scoring.
   Future work should incorporate longitudinal data and comorbidity scores
   to enrich the clustering signal beyond stay duration.
